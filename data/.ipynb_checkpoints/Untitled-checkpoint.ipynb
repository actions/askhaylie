```python
import pandas as pd
import numpy as np
plt.style.use('ggplot')
import seaborn as sns
from scipy import linalg
from sklearn import mixture
from sklearn.manifold import TSNE


import matplotlib.patheffects as PathEffects
import matplotlib as plt
%matplotlib inline
```

```python
ds=pd.read_csv("./nutrition.csv", index_col="no", header=None, skiprows=1,
names=["no","desc","water","energ","protein", "lipid_tot", "ash", "carb", "fiber", "sugar", "calcium", "iron", "magnesium","phosphorus", "potassium", "sodium", "zinc", "copper", "manganese", "selenium", "vit_c","thiamin", "riboflavin", "niacin", "panto_acid","vit_b6", "folate", "folic_acid", "food_folate", "folate_dfe", "choline_tot", "vit_B12", "vit_A_IU","vit_A_RAE","retinol", "alpha_carot", "beta_carot","beta_crypt", "lycopene", "lut_zea", "vit_E", "vit_D", "vit_D", "vit_K", "fa_sat", "fa_mono", "fa_poly", "cholestrl", "gmwt_1", "gmwt_desc1", "gmwt_2", "gmwt_desc2", "refuse_pct"]           
)
```

```python
ds.fillna(0.0, inplace=True)
ds.head(100)
```

```python
dpgmm = mixture.DPGMM(n_components=5, covariance_type='full')
```

```python
data = ds.drop(['desc', 'water' 'gmwt_1', 'gmwt_desc1','gmwt_2', 'gmwt_desc2', 'refuse_pct'], axis=1)
normalized = data[data["energ"]>0].div(data["energ"], axis=0).drop(["energ"], axis=1)
```

```python
normalized=normalized.dropna()
clusters = dpgmm.fit_predict(normalized)
```

```python
set(clusters)
dpgmm.score_samples(normalized)
```

```python
RS = 20150101
proj = TSNE(random_state=RS).fit_transform(normalized)
```
